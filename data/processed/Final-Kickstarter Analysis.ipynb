```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions  import date_format
from pyspark.sql.types import DateType
```

```python
spark = SparkSession.builder.appName('proposal_649').getOrCreate()
```

```python
df = spark.read.load("Kickstarter_2018-01-12T10_20_09_196Z.csv", inferSchema=True, header=True, format="csv",spt="\t")
```

```python
df.printSchema()
```

```python
df.toPandas()
```

```python
df.groupBy('disable_communication').count().show()
```

```python
df_project=df.withColumn('create_date',F.from_unixtime('created_at').cast(DateType())).withColumn('launch_date',F.from_unixtime('launched_at').cast(DateType()))
df_project=df_project.withColumn('deadline_at',F.from_unixtime('deadline').cast(DateType())).withColumn('state_change',F.from_unixtime('state_changed_at').cast(DateType()))
df_project.filter(df_project['state']=="successful").select('id','create_date',"deadline_at","state_change","state").show()
```

```python
df_project.toPandas()
```
